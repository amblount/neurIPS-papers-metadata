```
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from urllib.request import urlopen
url = "https://papers.nips.cc/book/advances-in-neural-information-processing-systems-32-2019"
print(type(url))
```

```
class CreateJSON:
  def __init__(self, url):
    self.html = urlopen(url).read()
    self.soup = BeautifulSoup(self.html)
    self.papers = self.collect_papers()
    self.prefix = 'https://papers.nips.cc/'
    self.json = {}

    print(self.soup.prettify())

  def collect_papers(self):
    papers = self.soup.find_all('li')

  def create_metadata(self):
    count = 0
    for paper in self.papers:
      #increase counter
      count += 1

      #create a unique id for each paper
      uni_id = 'paper' + str(count)

      #container for each papers metadata
      metadata = {}

      #collect the name of the paper
      name = paper.find('a').contents[0]
      #store the name in the metadata containter
      metadata['name'] = name
      
      #collect a partial link to the paper
      link = paper.find('a',href=True)
      #add link to the metadata container
      metadata['link'] = prefix + link['href']

      #gather author data
      a_data = self.create_author_hash(paper)

      #add author info to main hash
      metadata['authors'] = a_data 

      #add all metadata to json obj
      self.json[uni_id] = metadata

  def create_author_hash(paper):
    counter = 0

    #hash will contain all author data -> return object
    author_data = {}

    #find all authors on a paper
    authors = paper.find_all('a', attrs={'class', 'author'})

    #find the author page url on the NIPS site
    author_link = paper.find_all('a', attrs={'class', 'author'}, href=True)
    
    #iterate through each author on the paper
    for author in authors:

      #increase author count
      count += 1
    
      #create an author has for each author
      author__sub_data = {}

      #set the author name in the hash
      author_sub_data['author_name'] = author.contents

      #set author url in subhash
      author_sub_data['alink'] = self.prefix + author['href']
  
      #add author subdata to the author hash
      author_data["authornum" + str(count)] = author_data
  
papers = CreateJSON(url)
papers.collect_papers()
papers.create_metadata()
papers.json
```
